```python
import pandas as pd
```

```python
filename2109='/content/Mydrive/MyDrive/IOT_hackathon/Github/IOT_hackathon/data/서울특별시 공공자전거 이용정보(시간대별)_21.09.csv'
filename2110='/content/Mydrive/MyDrive/IOT_hackathon/Github/IOT_hackathon/data/서울특별시 공공자전거 이용정보(시간대별)_21.10.csv'
filename2111='/content/Mydrive/MyDrive/IOT_hackathon/Github/IOT_hackathon/data/서울특별시 공공자전거 이용정보(시간대별)_21.11.csv'
filename2112='/content/Mydrive/MyDrive/IOT_hackathon/Github/IOT_hackathon/data/서울특별시 공공자전거 이용정보(시간대별)_21.12.csv'
```

```python
data1=pd.read_csv(filename2109,encoding='cp949')
data2=pd.read_csv(filename2110,encoding='cp949')
data3=pd.read_csv(filename2111,encoding='cp949')
data4=pd.read_csv(filename2112,encoding='cp949')
```

```python
data=pd.concat([data1,data2,data3,data4])
```

```python
data.head()
```

```python
#연령대코드 더미변수 생성
data=data.replace('~10대','10대이하')
data=data.replace('70대~','70대이상')
연령대코드_type_dummy=pd.get_dummies(data['연령대코드'],prefix='연령')
```

```python
#성별 더미변수 생성
data['성별']=data['성별'].fillna('N')
data=data.replace('m','M')
data=data.replace('f','F')
성별_type_dummy=pd.get_dummies(data['성별'],prefix='성별')
```

```python
#필요없는 칼럼 제거
data=data.drop(['대여구분코드'],axis='columns')
data=data.drop(['성별'],axis='columns')
data=data.drop(['연령대코드'],axis='columns')
data=data.drop(['운동량'],axis='columns')
data=data.drop(['탄소량'],axis='columns')
```

```python
data=pd.concat([data,성별_type_dummy,연령대코드_type_dummy],axis=1)
```

```python
data=data.set_index("대여일자")
```

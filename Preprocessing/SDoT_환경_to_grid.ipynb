```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
sdot_env = pd.read_csv('Preprocessing/전처리완료/env_joined.csv')

sdot_env
```

```python
grid = gpd.read_file('Seoul_Grid/grid.shp')
area = gpd.GeoDataFrame.from_file('data/LSMD_ADM_SECT_UMD_11.shp', encoding='cp949')
```

```python
env_coord = sdot_env[['la', 'lo']]
env_coord
```

```python
gdf_env = gpd.GeoDataFrame(geometry = gpd.points_from_xy(x = env_coord.la, y= env_coord.lo, crs='WGS84'))
gdf_env = gdf_env.to_crs(epsg=5179)
```

```python
# S-DoT 환경 좌표처리
grid['sdot_env'] = 0
for i in tqdm(range(len(gdf_env))):
    idx = np.where(grid.contains(gdf_env.geometry[i]))
    if len(idx[0]) > 0 :
        grid.loc[grid.index==idx[0][0], 'sdot_env'] += 1
```

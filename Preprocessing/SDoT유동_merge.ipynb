```python
import pandas as pd
import numpy as np
import os
import requests
from tqdm import tqdm
```

```python
data_ls = os.listdir()
data_ls = sorted(data_ls)[1:]
```

```python
from pandas.io.parsers.python_parser import ParserError
df_whole = pd.DataFrame(columns = ['사이트명','방문자수','등록일자'])

for i in tqdm(range(len(data_ls))):
    try:
        df = pd.read_csv(data_ls[i], encoding='cp949',usecols=['사이트명','방문자수','등록일자'],low_memory=False)
    except ValueError:
        df = pd.read_csv(data_ls[i], encoding='cp949',usecols=['사이트명','방문자수 (명)','등록 일시'],low_memory=False)
        df.columns = ['사이트명','방문자수','등록일자']
    except ParserError:
        pass
    
    except TypeError:
        pass
    df.등록일자 = pd.to_datetime(df.등록일자, errors='coerce')
    df = df.groupby([pd.Grouper(freq='H', key='등록일자'), '사이트명']).mean().reset_index()
    df = df.astype({'사이트명':'int', '방문자수':'int'})
    df_whole = pd.concat([df_whole, df])

df_whole.sort_values(by=['사이트명','등록일자'])
df_whole.reset_index()
df_whole.to_csv('SDOT_population.csv')
```

```python
df_env = pd.DataFrame(columns = ['시리얼', '기온(℃)', '상대습도( %)', '미세먼지 보정(㎍/㎥)', '등록일자'])

for i in tqdm(range(len(data_ls))):
    try:
        df = pd.read_csv(data_ls[i], encoding='cp949', usecols=['시리얼', '기온(℃)', '상대습도( %)', '미세먼지 보정(㎍/㎥)', '등록일자'])
        df['등록일자'] = pd.to_datetime(df['등록일자'], errors='coerce')
        try:
            df = df.groupby([pd.Grouper(freq='H', key='등록일자'), '시리얼']).mean().reset_index()
            df_env = pd.concat([df_env, df])
        except:
            pass
    except Exception as e:
        print(e)
        df = pd.read_csv(data_ls[i], encoding='cp949', usecols=['시리얼', '기온(℃)', '상대습도(%)', '미세먼지 보정(㎍/㎥)', '등록 일자'])
        # df.rename(columns = {'상대습도(%)':'상대습도( %)', '등록 일자':'등록일자'}, inplace = True)
        df['등록일자'] = pd.to_datetime(df['등록 일자'], errors='coerce')
        try:
            df = df.groupby([pd.Grouper(freq='H', key='등록 일자'), '시리얼']).mean().reset_index()
    # df = df.astype({'시리얼':'int', '방문자수':'int'})
            df_env = pd.concat([df_env, df])
        except:
            pass
        
```

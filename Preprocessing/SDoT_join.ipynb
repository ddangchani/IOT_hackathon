```python
import pandas as pd
```

```python
pop_data = pd.read_csv('Preprocessing/전처리완료/SDOT_population.csv', index_col = 0)
pop_coord = pd.read_csv('Preprocessing/전처리완료/SDoT_pop_Coordinate.csv', index_col = 0)
```

```python
pop_coord
```

```python
pop_coord.head()
```

```python
pop_joined = pd.merge(pop_data, pop_coord, left_on='사이트명', right_on='modl_serial', how='left')

pop_joined.drop(columns = ['modl_serial', 'address'], inplace = True)
pop_joined.head()
```

```python
env_data = pd.read_csv('Preprocessing/전처리완료/SDOT_Env_pros_1.csv', index_col = 0)
env_coord = pd.read_csv('Preprocessing/전처리완료/SDoT_env_Coordinate.csv', index_col = 0)
```

```python
env_data.sort_values(by = '등록일자')
```

```python
env_coord.head()
```

```python
env_joined = pd.merge(env_data, env_coord, left_on='시리얼', right_on='modl_serial', how='left')

env_joined.drop(columns = ['modl_serial', 'address'], inplace = True)
env_joined.head()
```
